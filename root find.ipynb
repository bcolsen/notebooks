```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib notebook
f = lambda r: 2*r*np.sin(1001/(2*r)) - 1000
f_p = lambda r: - 2/r*np.cos(1001/(2*r)) + 2*np.sin(1001/(2*r))
new = lambda r: r - (f(r)/f_p(r))
n0 = 6000
iterations = 0
while True:
    n1 = new(n0)
    if abs(n1-n0) < 1e-13:
        break
    iterations += 1
    n0 = n1
r=np.linspace(4000, 10000, 1000)
plt.plot(r,f(r))
plt.axhline(0,color="C1")
plt.axvline(n0,color="C2")
print(n0, f(n0))
```

```python
%matplotlib notebook
h = lambda r: r-r*np.cos(1001/(2*r))
r=np.linspace(5000, 7000, 1000)
plt.plot(r,h(r))
plt.axvline(n0,color="C1")
plt.axhline(h(n0),color="C2")
h(n0)
```
