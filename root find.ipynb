```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib notebook
def f (r,s):
    return 2*r*np.sin(s/(2*r)) - 1000

def f_p (r,s):
    return - 2/r*np.cos(s/(2*r)) + 2*np.sin(s/(2*r))

def new (r,s):
    return r - (f(r,s)/f_p(r,s))
    
def do_new(r0,s):
    iterations = 0
    while True:
        r1 = new(r0,s)
        # print(interations, np.abs(r1-r0).max())
        if (np.abs(r1-r0) < 1e-10).all():
            return r1, iterations
        iterations += 1
        r0 = r1


r0 = 6000
r1, iterations = do_new(r0, 1001)
print(r1, f(r1,1001), iterations)
```

```python
%matplotlib notebook
r = np.linspace(4000, 10000, 1000)
plt.plot(r, f(r,1001))
plt.plot(r, f_p(r,1001))
plt.axhline(0, color="C1")
plt.axvline(r1, color="C2")
plt.xlabel("Track Radius (m)")
plt.ylabel("Origional Track Length - 1000 (m)")
```

```python
%matplotlib notebook
def h(r,s): 
    return r-r*np.cos(s/(2*r))

r=np.linspace(5000, 7000, 1000)
plt.plot(r,h(r,1001))
plt.axvline(r1,color="C1")
plt.axhline(h(r1,1001),color="C2")
plt.xlabel("Track Radius (m)")
plt.ylabel("Height (m)")
h(r1,1001)
```

```python
%matplotlib notebook
def hx(r0,x):
    s = x + 1000
    r, interations = do_new(r0,s)
    return r-r*np.cos(s/(2*r))

x = np.linspace(0.01, 10, 100)
r0 = np.ones_like(x) * 6000
plt.plot(x,hx(r0,x))
plt.axvline(1,color="C1")
plt.axhline(h(r1,1001),color="C2")
plt.xlabel("Extra Track (m)")
plt.ylabel("Height (m)")
```
